# FILLING NULL SPACES IN THE DATAFRAME
---

```python
Data.dropna().head(50)
```

```python
Data.describe()
```

### Deductions
---
For columns **fixed acidity**, **citric acid**, **density**, **pH**, **quality** and **chlorides**, we can replace the null values with the mean of the non null values since the values are between certain range. And for columns **free sulfur dioxide** and **total sulfur dioxide**we can use median since they are not in a particular or declared range. This is to ensure accurate results without any huge deviation from the actual values.

---

### Calculating The Mean, Mode and Median Of The Columns With Null Values
---

```python
mean_FA = Data.fixed_acidity.mean()
mean_CA = Data.citric_acid.mean()
mean_Chl = Data.chlorides.mean()
median_FSD = Data.free_sulfur_dioxide.median()
median_TSD = Data.total_sulfur_dioxide.median()
mean_Dsy = Data.density.mean()
mean_pH = Data.pH.mean()
mean_Q = Data.quality.mean()
```

```python
print(f'The mean of Fixed Acidity = {mean_FA}')
print(f'The mean of Citric Acid = {mean_CA}')
print(f'The mean of Chlorides = {mean_Chl}')
print(f'The median of Free Sulfur Dioxide = {median_FSD}')
print(f'The median of Total Sulfur Dioxide = {median_TSD}')
print(f'The mean of Density = {mean_Dsy}')
print(f'The mean of pH = {mean_pH}')
print(f'The mean of Quality = {mean_Q}')
```

### Filling The Null Spaces In The DataFrame with The Above Info
---

```python
Data.fixed_acidity = Data.fixed_acidity.fillna(mean_FA)
Data.citric_acid = Data.citric_acid.fillna(mean_CA)
Data.chlorides = Data.chlorides.fillna(mean_Chl)
Data.free_sulfur_dioxide = Data.free_sulfur_dioxide.fillna(median_FSD)
Data.total_sulfur_dioxide = Data.total_sulfur_dioxide.fillna(median_TSD)
Data.density = Data.density.fillna(mean_Dsy)
Data.pH = Data.pH.fillna(mean_pH)
Data.quality = Data.quality.fillna(mean_Q)
```

## QUALITY
---

```python
x = Data.quality < mean_Q
```

```python
x.value_counts()
```

```python
plt.pie(x.value_counts(), labels = ['False','True'])
plt.legend(x.value_counts())
plt.title('CORRELATION BETWEEN THE QUALITY VALUES AND ITS MEAN')
```

### Deduction
This pie chart represents the quality rates of all the samples (1601 entries). This shows the correlation between the quality values of the entries and their mean. The False (Blue) part represents the values greater than the mean value and the True (Orange) represents the values less than the mean value. Therefore we can deduce that more quality red wine are produced.
